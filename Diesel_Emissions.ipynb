<center> <h1> Kompresijas aizdedzes motoru kaitīgie izmeši

<center> <h2> Kaitīgo izmešu veids un nozīmīgums

Kompresijas aizdedzes motoru izplūdes gāzēs ir galvenie kaitīgo izmešu veidi ir:
- oglekļa dioksīds ($CO_2$);
- slāpekļa oksīdi ($NO_x$);
- nesadegušie ogļūdeņraži ($HC$);
- cietās daļiņas jeb kvēpi ($PM$). <br>

$CO_2$ izmešu daudzums ir mazāks, nekā dzirksteļaizdedzes motoriem, jo kompresijas aizdedzes motori darbojas ar augstāku lietderību, un to īpatnējais degvielas patēriņš (uz jaudas vienību vai automobiļa nobraukto attālumu) ir mazāks. 
Šis izmešu veids ir kaitīgs dabai, bet nav tiešā veidā toksisks.<br> <br>
$NO_x$ koncentrācija motora izplūdes gāzēs ir līdzīga kā dzirksteļaizdedzes motoriem, bet automobiļa izpūtēja izejā tā parasti ir augstāka.  Kompresijas aizdedzes motoru izplūdes sistāmā ir grūtības pārveidot un samazināt šī izmešu veida daudzumu. $NO_x$ izmeši ir kaitīgi gan tiešā veidā, gan kā ozona un citu kairinošo un kaitīgo vielu prekursori. Motoros pārsvarā veidojas slāpekļa oksīds $NO$, bet dabai kaitīgs ir slāpekļa dioksīds $NO_2$, kas izveidojas no $NO$ vēlāk,  atmosfērā. Savukārt $NO_x$ daudzumu uzskaita kā  $NO$ un  $NO_2$ kopējo daudzumu, kas parasti tiek pārrēķināts $NO_2$ vienībās.  <br> <br>
Nesadegušo ogļūdeņŗažu koncentrācija kompresijas aizdedzes motora izplūdes gāzēs ir zemāka nekā dzirksteļaizdedzes motoriem, bet pietiekami nozīmīga un kaitīga dabai. $HC$ izmeši rada dīzeļmotoru izmešiem raksturīgo smaku, kā arī baltos dūmus auksta motora iedarbināšanas laikā.<br><br>
Cietās daļiņas ir raksturīga un nozīmīga kompresijas aizdedzes motoru izmešu komponente. To izmērs ir 10...500 nm izmērā. Cietās daļiņas pārsvarā sastāv no kvēpiem un tajos absorbētiem ogļūdeņražiem. Apmēram 0.2...0.5 % no degvielas masas tiek izvadīti cieto daļiņu veidā.




<center> <h2> Kaitīgo izmešu veidošanās sadedzes procesā

Kaitīgo izmešu veidošanās mehāniskmi tiešās iesmidzināšanas kompresijas aizdedzes (KI) motoros parādīti 1. attēlā.



<div align="center">  1. att. Kaitīgo izmešu veidošanās mehāniskmi KI motoros. [Heywood 2018]  <br>
<div align="left"> <br>
    
    
Iepriekšsajauktās sadedzes fāzē (Premixed), kas seko pēc aizdegšanās kavējuma, galvenokārt veidojas $HC$ un $PM$. <br>
Šīs sākotnējās fāzes laikā gāzu temperatūra nav pietiekami augsta, lai veidotos $NO$. Savukārt strūklas ārējās zonās, kur degmaisījums ir liess, sadegšana notiek nepilnīgi, un veidojas $HC$. Degvielas tvaiku strūklas iekšpusē notiek degvielas pirolīze un kvēpu veidošanās. <br>
    
Difūzijas sadedzes fāzē liesma ir pietiekami karsta un atrodas zonā, kas bagāta ar skābekli. Rezultātā veidojas slāpekļa oksīds $NO$. $HC$ izmešu veidošanās galvenie iemesli ir divi: liesmas nodzišana pie sadegšanas kameras sienām un degvielas tvaiki no sprauslas gala nodalījuma. Tāpat degvielas strūklas iekšpusē veidojas kvēpi, kas pēc tam daļēji spēj oksidēties.
    

<center> <h2> $NO_x$ izmešu veidošanās kompresijas aizdedzes motoros

<center> <h3> $NO$ izmešu veidošanās kinētika

Motora degkamerā sākotnēji veidojas slāpekļa oksīds $NO$. <br>
Galvenās ķīmiskās rekacijas šajā procesā ir trīs:
$$ N_2+O=NO+N$$
$$ N+O_2=NO+O$$
$$ N+OH=NO+H$$
<br> Šīs reakcijas notiek gan uz priekšu, gan atpakaļ, bet reakciju virziena ātrums ir atšķirīgs un tā atkarība no temperatūras ir mainīga. <br> Ja temperatūra degkamerā samazinātos lēni un vienmērīgi, daļa no $NO$ pārveidotos atpakaļ par $N_2$ un $O_2$. Sadedzes procesa noslēgumā degkamerā palielinās tilpums, gāzes temperatūra degkamerā samazinās, ķīmiskais līdzsvars tiek izjaukts, un rezultātā sadedzes produktos paliek ievērojams $NO$ daudzums. Reakciju ātruma konstantes ir iepriekš labi izpētītas un publicētas. *Heywood 2018* grāmatā, 607. lpp ir parādīts, kā var modelēt $NO$ veidošanos motora degkamerā.

<center> <h3> $NO_2$ izmešu veidošanās kinētika

Analizējot slāpekļa oksīdu veidošanos ķīmisko kinētiku, šķiet, ka $NO_2$ daudzumam motora izmešos vajadzētu būt niecīgam. Tas tā tiešām ir dzirksteļaizdedzes motoros, bet ne KA motoru gadījumā. Eksperimentāli analizējot KA motoru izmešus, tajos $NO_2$ daudzums ir 10...30% no kopējā $NO_x$ daudzuma. Galvenās ķīmiskās reakcijas attiecībā un $NO_2$ veidošanos ir sekojošas. $NO_2$ veidojas liesmas zonā:
$$NO+HO_2 \rightarrow NO_2+OH$$
un pārveidojas atpakaļ par $NO$:
$$NO_2+O \rightarrow NO+O_2$$
Atgriezenisko reakciju var izjaukt, ja gāzes, kurā izveidojies $NO_2$ tiek strauji atdzesēta, sajaucot to ar vēsāku gāzi. Šāds process notiek KI motoru degkamerā, un rezultātā izmešos ir salīdzinoši daudz $NO_2$. <br>
$NO_2$ relatīvais daudzums KA motora $NO_x$ izmešos un motora slodzes ietekme parādīta 2. att.

<div align="center">  2. att. $\frac{NO_2 }{ NO_x}$ attiecība kā funkcija no apgriezieniem un slodzes KA motoros. [Heywood 2018]  <br>
<div align="left"> <br>
Jo mazāka slodze, jo mazāka degvielas porcija ir ievadīta sadegšanas kamerā. Lielāks daudzums vēsas gāzes rada apstākļus, kuros izveidojas un atgāzēs paliek vairāk $NO_2$. <br> Tāpat, jo lēnāki ir motora apgriezieni, jo lielāks $NO_2$ īpatsvars izmešos. 

<center> <h3> $NO$ izmešu veidošanās kompresijas aizdedzes motoru degkamerā

Iepriekšsajauktas degšanas fāzē $NO$ izmeši faktiski neveidojas. Toties difūzijas kontrolētājā degšanas fāzē tie veidojas ievērojamā daudzumā. 3. attēlā parādīta $NO$ izmešu veidošanās intensitāte difūzijas liesmas zonas tuvumā. Attēla kreisajā pusē ir degvielas tvaiku strūklas puse, kur degmaisījums ir trekns. Savukārt labajā pusē ir degkamerā esošā saspiestā gāze (gaisa un atliku gāzu maisījums).


<div align="center">  3. att. $NO$ veidošanās shēma ap degvielas tvaiku strūklu. [Heywood 2018]  <br>
<div align="left"> <br>
 
Galvenokārt $NO$ veidojas liesmas zonas liesajā pusē.

Analizējot dažādu gāzes komponenšu koncentrācijas izmaiņas, kā arī spiediena un temperatūras izmaiņas gāzu paraugu ieguves punktā "S", izveidota 4. att redzamā diagramma.


<div align="center">  4. att. Apstākļi un gāzes veidu koncentrācija sadegšanas kamerā. [Heywood 2018]  <br>
<div align="left"> <br>
 Redzams, ka sadedzes laikā $NO$ koncentrācija strauji palielinās, bet tad sāk samazināties. Galvenais šīs parādības iemesls ir  gāzu sajaukšanās degkamerā, kad sadedzes process ir beidzies. Šie dati sniedz interasntu ieskatu sadedzes procesa norisē, bet tikai vienā degkameras punktā.

Nākošajā, 5. attēlā parādīta vidējās $NO$ koncentrācijas izmaiņas degkamerā. Šeit var ieraudzīt, ka $NO$ sadalīšanās beidzoties sadedzes procesam notiek, bet efekts ir diezgan nenozīmīgs.


<div align="center">  5. att. Vidējās $NO$ koncentrācijas degkamerā attiecība pret koncentrāciju izplūdes sistēmā. [Heywood 2018]  <br>
<div align="left"> <br>
    
Tā kā parādīta ir koncentrāciju attiecība, var novērtēt, ka motora apgriezienu izmaiņas rada niecīgu efektu uz $NO$ veidošanās un sadalīšanās līkni.

<center> <h3> $NO$ izmešu veidošanās samazināšana motoru degkamerā

Viens no efektīvākajiem veidiem, kā samazināt $NO$ veidošanos degamerā, ir tā dēvēto atšķaidītgāzu lietošana. <br> Atšķaidītgāzi pievieno ieplūdes gaisam. Eksperimenti ar augstas un zemas siltumietilpības gāzu (piemēram $CO_2$ un $N_2$) pievienošanu gaisam rāda, ka galvenais efekts veidojas samazinot tieši samazinot degošās gāzes temperatūru liesmas zonās, ko sasniedz, samazinot skābekļa koncentrāciju. <br> Vidējās gāzes temperatūras samazināšanai cilindrā ir mazāka nozīme.
Lēts un salīdzinoši vienkārš paņēmiens ir motorā ieplūstošā gaisa atšķaidīšana ar izplūdes gāzēm, lietojot izplūdes gāzu recirkulācijas jeb EGR sistēmu. <br> 6. att. parādīta motora slodzes un izplūdes gāzu koncentrācijas pieplūstošajā gaisā ietekme uz $NO$ koncentrāciju.


<div align="center">  6. att. Motora slodzes un izplūdes gāzu koncentrācijas ietekme uz $NO$ koncentrāciju. [Heywood 2018]  <br>
<div align="left"> <br>
    
Redzmas, ka KA motorā pie zemas slodzes pat 50% no ieplūstošā gaisa var aizvietot ar izplūdes gāzēm, panākot ievērojamu $NO$ izmešu daudzuma samazinājumu. Jāņem vērā tād KA motoru īpatnība, ka pie zemas slodzes tiek iesmidzināts neliels degvielas daudzums, un izplūdes gāzēs ir maza sadegšanas produktu koncentrācija. Tas ir diezgan atšķirīgi, salīzinot ar dzirskteļaizdedzes motoriem.


<center> <h2> $HC$ izmešu veidošanās kompresijas aizdedzes motoros

Dīzeļdegviela satur ogļūdeņražus, kam ir lielākas molekulārais svars, un attiecīgi, tiem ir augstāka viršanas temperatūra, salīdzinot ar benzīnu. Sadedzes procesā notiek degvielas komponenšu pirolīze. Kopumā KA motoru $HC$ izmešu sastāvs ir sarežģītāks un satur lielākas molekulas, nekā līdzīga rakstura izmeši no dzirksteļaizdedzes motora. <br>

$HC$ izmešu veidošanās abās KA motoram raksturīgajās sadedzes fāzēs ir parādīta 7. attēlā.


<div align="center">  7. att. $HC$ veidošanās mehānisms KI motoram raksturīgā sadedzes procesā [Heywood 2018]  <br>
<div align="left"> <br>
    
Degviela, kas izsmidzināta aizdegšanās kavējuma periodā (7. att. a), iztavaiko un katrā degvielas tvaiku strūklā sajaucas ar apkārtējo gaisu. Strūklas ārējās zonās izveidojas degmaisījums, kas ir tik liess, ka tā pašaizdegšanās esošajos apstākļos nenotiek. Lielākā daļa no šī liesā degmaisījuma, iespējams, sadegs motora cikla turpinājumā. Eksperimenti rāda, ka tieši šis $HC$ izmešu avots KA motoru gadījumā ir visnozīmīgākais. <br>
    
Savukārt galvenajā, difūzijas kontrolētājā sadedzes fāzē (7. att. b), daļa no nepilnīgi sadegušā degmaisījuma pārāk lēni sajaucas ar apkārtējo gaisu degkamerā, un liesma nodziest. Šis mehānisms arī rada daļu no $HC$ izmešiem.
    


<center> <h3> $HC$ izmešu veidošanās pārāk liesā degmaisījumā

Iesmidzinātā degviela cilindrā iztvaiko un sajaucas ar gaisu, veidojot apgabalus ar atšķirīgu gaisa/degvielas attiecību. Degvielas strūklas iekšpusē ir izteikts degvielas pārums, degmaisījums ir trekns. Savukārt strūklas perifērijā veidojas apgabali, kuros degmaisījums ir liesāks par aizdegšanās robežu ($\lambda$ > 1.3). Šādu liesa degmaisījuma apgabalu apjoms kļust arvien lielāks aizdegšanās kavējuma perioda laikā. Kad sākas sadedzes, šajos apgabalos degvielas oksidēšanās notiek lēni un nepilnīgi, veidojoties daļējas sadegšanas produktiem (tie aldehīdi un citi skābekli saturošie savienojumi). <br>
$HC$ izmešu daudzums, kas veidojas no sadedzes šajos liesā degmaisījuma apgabalos, ir atkarīgs no iesmidzinātās degvielas aizdegšanās kavējuma periodā, sajaukšanās intensitātes šajā periodā un apstākļiem motora degkamerā pašaizdedzes brīdī. 


<div align="center">  8. att. Korelācija starp $HC$ izmešiem izplūdē un aizdegšanās kavējumu [Heywood 2018]  <br>
<div align="left"> <br>
    
8. att. parādīta sakarība starp aizdegšanās kavējumu un $HC$ izmešiem. Jo ilgāks kavējums, jo lielāka izmešu koncentrācija.

<center> <h3> $HC$ izmešu veidošanās nepietiekami sajauktā degmaisījumā

Nepietiekami sajaukts degmaisījums, kas rada $HC$ izmešus, var izveidoties divos atšķirīgos apstākļos. <br> <br>

KA motoros, melni dūmi izplūdē pilnas slodzes stabilā darba režīmā ("pedālis grīdā") parādās, kad vidējā gaisa-degvielas attiecība ir apmērmam 1.3. Savukārt pārejas režīmos (paātrinājums) melni dūmi izplūdē parādās, ja notiek īslaicīga  nepieciešamā degvielas daudzuma pārsniegšana. Nelielos degvielas strūklas apgabalos izveidojas lokāli pārāk trekni apgabali, kas saglabājas izpletes un izplūdes procesos. $HC$ izmešu daudzums tādos apstākļos strauji pieaug, un šis efekts ir novērojams motora paātrinājuma režīmos. <br> <br>

Otrs šāda veida $HC$ izmešu avots ir degviela, kas degkemerā nonāk no smidzinātāja papildus nodalījuma. Smidzinātāja galā ir izveidots papildus nodalījums, kas nepieciešams spiediena izlīdzināšanai. Dažādu smidzinātāju šķērsgriezums parādīts 9. att.


<div align="center">  9. att. Smidzinātāja šķērsgriezums un korelācija starp $HC$ izmešiem izplūdē un smidzinātāja nodalījuma tilpumu [Heywood 2018]  <br>
<div align="left"> <br>

Pēc tam, kad degvielas iesmidzināšana ir pabeigta, izpletes procesa laikā, degviela šajā nodalījumā iztvaiko vai no tā izpil, un relatīvi lēni iztvaiko un sajaucas ar gaisu. Šie degviela var sadegt nepilnīgi un radīt $HC$ izmešus. <br>
Korelācija starp $HC$ koncentrācju izplūdē un smidzinātāja nodalījuma tilpumu parādīta 9. att.
<br> <br>
Mūsdienu KA motoros parasti lieto varākkārtīgu iesmidzināšanu. Šis motora vadības paņēmiens uzlabo vairākus motora darbības aspektus, bet var palielināt tā $HC$ izmešus.




<center> <h3> $HC$ izmešu veidošanās liesmai nodziestot

KA motoru $HC$ izmešu daudzums ir jutīgs attiecībā uz motora eļļas un dzesēšanas šķidruma temperatūru. Zemā motora darba temperatūrā liesma tuvumumā degkameras sienām nodziest. Izmešu daudzums ir atkarīgs arī no strūklas pieskaršanās degkameras sienām. <br> Savukārt pilnīga degmaisījuma neizdegšanās, kas mēdz regulāri gadīties dzirksteļaizdedzes motoros, KA motoros parasti nenotiek. 

<center> <h2> Cieto daļiņu veidošanās

Cieto daļiņu jeb kvēpu izmeši ir izteikta dīzeļmotoru darbības iezīme. Cietās daļiņas veidojas degvielas strūklā. Pirolīze galvenais cieto daļiņu veidošanās veicinātājs. Degvielas termiskā pirolīze notiek nepietiekama skābekļa daudzuma apstākļos, temperatūrā, kas augstāka pa par 670 *K*. Degvielas molekulas sadalās, izveidojot ūdeņradi $H$, hidoksīda $OH$ un metila $CH_3$ radikāļus un dažādus citus savienojumu, piemēram etilēnu $C_2H_4$ un acetilēnu $C_2H_2$. Tad šīs jaunās vielas veido polickliskos aromātiskos savienojumus ($PAH$). Šis process parādīts 10. att.<br>



<div align="center">  10. att. $PAH$ veidošanās shēma [Raza et al. 2018]  <br>
<div align="left"> <br>

$PAH$ veidošanās reakcijas ir reversīvas, un temperatūrā, kas augstāka par 1800 *K*, $PAH$ sadalās. <br> <br> $PAH$ molekulām ir tendence apvienoties klasteros. Šie klasteri savukārt reaģē ar acetilēnu, un šajā procesā izveidojas kvēpu daļiņas. 
Kvēpu veidošanos dēvē par *nukleāciju* un daļiņas par *nukleācijas režīma* daļiņām. Tādā veidā izveidojas daļiņas ar izmēru, kas mazāks 50 *nm*. Lielākā daļa šo daļiņu sadeg motora degkamerā.
    
Kvēpu veidošanos iepriekš sajauktas un difūzijas kontrolētas degšanas fāzēs parādīta 11. att.
    
    
<div align="center">  11. att. Cieto daļiņu veidošanās dažādās sadedzes fāzēs [Heywood 2018]  <br>
<div align="left"> <br>

Sākotnēji daļiņu veidošanās notiek *nukleācijas* režīmā. Tālākā daļiņu attīstība notiek *akumulācijas* režīmā. Sīkās kvēpu daļiņas uzkrājas un apvienojas, veidojot ievērojami lielākas saliktas cietās daļiņās. *Akumulācijas* režīmā izveidojas daļiņas ar izmēru 50...500 *nm*.<br> Kopējā daļiņu veidošanās shēma ir parādīta 12. att. <br>
11. att. redzams, ka lielākā daļiņu koncentrācja ir difūzijas kontrolētājā sadedzes fāzē. Tās atrodas degvielas strūklas iekšpusē, aptvertas ar difūzijas liesmu. Daļiņas sakarst un sāk kvēlot. Nonākot strūklas ārējās zonās, kur ir pietiekami daudz skābekļa, liela daļa cieto daļiņu oksidējas.

<center> <h2> Kopsavilkums

Lekcijas konspektā apskatīta slāpekļa oksīdu $NO_x$, ogļūdeņražu $HC$ un cieto daļiņu $PM$ izmešu veidošanās kompresijas aizdedzes motoros. $NO_x$ un $PM$ izmeši galvenokārt veidojas difūzijas kontrolētajā sadedzes fāzē. Savukārt galvenais $HC$ izmešu avots ir iepriekšsajauktās sadedzes process. <br> Tā kā lielākā kaitīgo izmešu daļa ir tieši $NO_x$un $PM$, arī motoru attīstība ar mērķi samazināt tā darbības ietekmi uz apkārtējo vidi saistās ar šīs sadedzes fāzes norises uzlabošanu un vai pat pilnīgu izslēgšanu no sadedzes procesa. <br> Tas pilnībā vai daļēji tiek realizēts *HCCI* un *PPC* veida kompresijas aizdedzes motoros.
